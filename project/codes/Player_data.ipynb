```python
import pandas as pd
from pandas import DataFrame
import csv
```

```python
data_dir = "../data/player_Data/players_22.csv"      # 2021/22 시즌 EPL 선수 데이터 경로
fb_data = pd.read_csv(data_dir, encoding='utf-8')
```

```python
EPLteams = ['Arsenal', 'Aston Villa', 'Brentford', 'Brighton & Hove Albion', 'Burnley', 'Chelsea', 'Crystal Palace',
            'Everton', 'Leeds United', 'Leicester City', 'Liverpool', 'Manchester City', 'Manchester United',
            'Newcastle United', 'Norwich City', 'Southampton', 'Tottenham Hotspur', 'Watford', 'West Ham United', 'Wolverhampton Wanderers']
EPL = {}                    # EPL 팀 별 선수 목록 : EPL 딕셔너리 : EPL[ 팀명 ] : [ [선수1], [선수2] ....]
```

# epl 팀별 분류

```python
result = []
for team in EPLteams:
    EPL[team] = []
    print(team)
    for i, row in fb_data.iterrows():
        if row['club_name'] == team:
            EPL[team].append(fb_data.loc[i])
```

```python
for team in EPLteams:
    col = EPL[team][0].keys()
    value_items = [] 
    with open("{0}.csv".format(team),'w+',encoding='utf-8-sig') as f:
        wr = csv.writer(f,quoting=csv.QUOTE_ALL)
        wr.writerow(col)        
        f.close()
        
```

```python
for team in EPLteams:
    with open("{0}.csv".format(team),'a',encoding='utf-8-sig') as f:
        for overall_player_data in EPL[team]:
            value_items = []
            for col in overall_player_data:
                if(type(col) != 'str'):
                    col = str(col)
                else:
                    pass
                value_items.append(col)
                
            wr = csv.writer(f,quoting=csv.QUOTE_ALL)
            wr.writerow(value_items)        
    f.close()
```
